Size

```python
ax=sns.countplot(x='Size', data=df)
```

```python
ax=sns.countplot(x='Size', data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

Note: From above Graph you can see that most of the people buys M size


Group By

```python
df.groupby(['Size'], as_index=False)['Quantity'].sum().sort_values(by='Quantity', ascending=False)
```

```python
S_Quantity=df.groupby(['Size'], as_index=False)['Quantity'].sum().sort_values(by='Quantity', ascending=False)
sns.barplot(x='Size', y='Quantity', data=S_Quantity)
```

Note: From the above graph you can see that most of the Quantity buys M-Sizein the sales

Courier Status

```python
sns.countplot(data=df, x='Courier Status', hue='Courier Status')
```

```python
#histogram
df['Size'].hist()
```

```python
df['Category']= df['Category'].astype('str')
```

```python
column_data= df['Category']
plt.figure(figsize=(10,5))
plt.hist(column_data, bins=10, edgecolor='black')
plt.xticks(rotation=90)
plt.show()
```

Note: From the above graph you can see that most of the buyers are of T-shirt

```python
#Checking B2B data by using pie chart
B2B_check = df['B2B'].value_counts()

#Plot the pie chart
plt.pie(B2B_check, labels=B2B_check.index, autopct='%1.2f%%')


#plt.axis('equal')
plt.show()
```

Note: From the above chart we can see the maximum i.e. 99.3% of buyers are retailers and 0.8% are B2B buyers

```python
#Prepare the data for scatter plot
x_data = df['Category']
y_data = df['Size']
```

```python
#Plot the scatter graph
plt.scatter(x_data, y_data)
plt.xlabel('Category')
plt.ylabel('Size')
plt.title('Scatter plot')
plt.show()
```

```python
#Plot count of cities by state
plt.figure(figsize=(12,6))
sns.countplot(data=df, x='ship-state')
plt.xlabel('ship-state')
plt.ylabel('Count')
plt.title('Distribution of State')
plt.xticks(rotation=90)
plt.show()
```

```python
#Top 10 states
top_10_state = df['ship-state'].value_counts().head(10)

#Plot count of cities by state
plt.figure(figsize=(12,6))
sns.countplot(data=df[df['ship-state']. isin(top_10_state.index)], x='ship-state')
plt.xlabel('ship-state')
plt.ylabel('Count')
plt.title('Distribution of State')
plt.xticks(rotation=45)
plt.show()
```

Note: From the above graph you can see that most of the buyers are from Maharashtra state

Conclusion

The data analysis reveals that the business has a significant customer base in Maharashtra state, mainly serves retailers, fulfills orders through Amazon, experiences high demand for T-shirts, and sees M-size as the preferred choice among buyers.
